### EXTRACTING PIXEL VALUES AND COLOR SPACE OF THE IMAGE

### ROTATING THE IMAGE

```python
import cv2  # Import the OpenCV library for computer vision tasks
from matplotlib import pyplot as plt  # Import Matplotlib's pyplot for plotting

# Load the image
img = cv2.imread('image.png')  # Read the image from a file and store it in 'img'

# Create the rotation matrix
(height, width) = img.shape[:2]  # Get the dimensions of the image
center = (width / 2, height / 2)  # Find the center point of the image
matrix = cv2.getRotationMatrix2D(center, 90, 1.0)  # Define the rotation matrix for a 90-degree rotation

# Rotate the image
rotated = cv2.warpAffine(img, matrix, (width, height))  # Apply the rotation to the image

# Display the rotated image using Matplotlib
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))  # Convert the color space from BGR to RGB
plt.title('Rotated Image')  # Set the title of the plot
plt.show()  # Show the image in a Matplotlib window
```

This Python code is written to rotate an image by 90 degrees. The purpose of the code is to load an image using the OpenCV library, create a rotation matrix to rotate the image around its center by 90 degrees, and apply this transformation. The rotated image is then displayed with Matplotlib after converting the color space from BGR to RGB. This process is a commonly used method in the field of image processing and computer vision and is a fundamental technique for performing geometric transformations on images. For example, this piece of code can be used as a preprocessing step in image recognition or object tracking systems.

### CONVERTING THE IMAGE TO GRAYSCALE AND DETECTING EDGES

```python
import cv2
from matplotlib import pyplot as plt

# Load the image
img = cv2.imread('image.png')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect edges
edges = cv2.Canny(gray, 30, 100)  # minVal = 30, maxVal = 100

# # Display the edges
# cv2.imshow('Edge Image', edges)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Display the image
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)) #BGR TO RGB
plt.title('Edges')
plt.show()
```

This script uses OpenCV and Matplotlib libraries to process and display an image. First, it loads an image named image.png using OpenCV. The image is then converted to grayscale to simplify the edge detection process. Using the Canny edge detection algorithm, edges within the grayscale image are identified with threshold values set to 30 and 100. Although the code to display the edge-detected image using OpenCVs imshow function is commented out, the script instead uses Matplotlib to display the edges. The image is converted from BGR to RGB color space before being displayed with the title Edges.

### FOURIER TRANSFORM

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image and convert it to grayscale
img = cv2.imread('image.png', 0)

# Compute the Fourier transform
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)

# Display the spectrum
magnitude_spectrum = 20*np.log(np.abs(fshift))
plt.subplot(121), plt.imshow(img, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap = 'gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

This Python script uses the OpenCV and NumPy libraries to perform a Fourier Transform on an image and visualize its frequency spectrum. The script starts by loading an image in grayscale. It then computes the 2D Fourier Transform of the image using np.fft.fft2, which converts the image from the spatial domain to the frequency domain. The result is shifted to center the low frequencies using np.fft.fftshift. The magnitude spectrum of the transformed image is calculated and displayed using Matplotlib. The original grayscale image and its corresponding magnitude spectrum are shown side by side for comparison. The magnitude spectrum reveals the frequency components present in the image, with brighter spots indicating higher magnitude frequencies.

The magnitude spectrum is a type of graph that shows the magnitudes of different frequencies in a signal or image. The spectrum appears as a grayscale image with variable intensities, where brighter areas represent higher magnitudes (or powers) of the corresponding frequencies. The bright vertical line in the center indicates a significant amount of energy at a particular frequency. This type of spectrum is important in fields such as signal processing and image analysis, as it helps to understand the frequency content of signals or images. This can be important for tasks such as filtering, compression, or enhancement.

### HOUGH TRANSFORM FOR LINE DETECTION

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('image.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Perform edge detection
edges = cv2.Canny(gray, 100, 200, apertureSize=3)

# Detect lines with Hough transform
lines = cv2.HoughLines(edges, 1, np.pi / 180, 250)

# Draw the detected lines on the image
if lines is not None:
    for rho, theta in lines[:, 0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)

# Display the image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Detected Lines')
plt.show()
```

This script demonstrates how to use the Hough Transform for line detection in an image using OpenCV. The process begins by loading an image and converting it to grayscale. Edge detection is then performed on the grayscale image using the Canny edge detector. The Hough Transform is applied to the edges to detect lines, which are represented by their polar coordinates (rho and theta). For each detected line, the script calculates the endpoints and draws the lines on the original image in red. Finally, the image with the detected lines is displayed using Matplotlib.

### HOUGH TRANSFORM FOR CIRCLE DETECTION

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('image.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.medianBlur(gray, 5)

circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 1, 20,
                            param1=60, param2=40, minRadius=10, maxRadius=50)


# Draw the detected circles on the image
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        # Draw the center of the circle
        cv2.circle(image, (i[0], i[1]), 1, (0, 100, 100), 3)
        # Draw the circle itself
        cv2.circle(image, (i[0], i[1]), i[2], (255, 0, 255), 3)

# Display the image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) #BGR TO RGB
plt.title('Detected Circles')
plt.show()
```

This Python script uses the OpenCV library to detect circles in an image using the Hough Transform method. The process begins by loading an image and converting it to grayscale. A median blur is then applied to reduce noise. The cv2.HoughCircles function is used to detect circles in the processed image, with specific parameters set for edge detection and circle detection sensitivity. If circles are detected, they are drawn on the original image: the center of each circle is marked with a small circle, and the circumference is drawn with a larger circle. Finally, the image with the detected circles is displayed using Matplotlib, with the color space converted from BGR to RGB for correct color representation.

### MORPHOLOGICAL TRANSFORMATIONS

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image and convert it to grayscale
img = cv2.imread('image.png', 0)

# Dilation
kernel = np.ones((5,5),np.uint8)
dilation = cv2.dilate(img, kernel, iterations = 1)

# # Display the dilated image
# cv2.imshow('Dilated Image', dilation)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(dilation, cv2.COLOR_BGR2RGB)) #BGR TO RGB
plt.title('Dilated Image')
plt.show()

# Erosion
erosion = cv2.erode(img, kernel, iterations = 1)

# # Display the eroded image
# cv2.imshow('Eroded Image', erosion)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Display the image
plt.imshow(cv2.cvtColor(erosion, cv2.COLOR_BGR2RGB)) #BGR TO RGB
plt.title('Eroded Image')
plt.show()
```

This script demonstrates morphological transformations using OpenCV in Python. It starts by loading an image and converting it to grayscale. The first transformation applied is dilation, which uses a 5x5 kernel of ones. Dilation adds pixels to the boundaries of objects in an image, making them more prominent. The dilated image is then displayed using Matplotlib. Next, the script applies erosion using the same kernel. Erosion removes pixels on object boundaries, effectively shrinking the objects. The eroded image is also displayed using Matplotlib. Both transformations help in various image processing tasks, such as noise removal and object separation.

### TRESHOLDING

```python
import cv2
from matplotlib import pyplot as plt

# Load the image and convert it to grayscale
img = cv2.imread('image.png', 0)

# Thresholding
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# cv2.imshow('Thresholded Image', thresh)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB)) #BGR TO RGB
plt.title('Thresholded Image')
plt.show()
```

This Python script uses the OpenCV and Matplotlib libraries to process and display an image. First, it loads an image named image.png and converts it to grayscale. Then, it applies a binary threshold to the grayscale image, where pixel values above 127 are set to 255 (white), and those below are set to 0 (black). Instead, the script uses Matplotlib to display the thresholded image, converting the color format from BGR to RGB for correct color representation. ## Image Resizing and Comparison Display

```python
import cv2  # Import the OpenCV library for computer vision tasks
from matplotlib import pyplot as plt  # Import Matplotlib's pyplot for plotting

# Load the original image
original_image = cv2.imread("image.png")

# Get the dimensions of the image
h, w = original_image.shape[:2]

# Resize the image
resized_image = cv2.resize(original_image, (w*2, h*2))  # Double the width and height

# Display both the original and resized images
plt.figure(figsize=(16, 8))  # Set the figure size

# Display the original image
plt.subplot(1, 3, 1)  # Position it in the first cell of a 1 row by 3 columns grid
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

# Display the resized image in a larger subplot
plt.subplot(1, 3, (2, 3))  # Span it across the second and third cells of the grid
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title('Resized Image - Larger View')

# Start the display process
plt.show()
```
